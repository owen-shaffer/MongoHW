```python
action_movie = collection.find_one({"genres": "Action"})


pprint.pprint(action_movie)
```

```python
movies_after_2000 = list(collection.find({"year": {"$gt": 2000}}).limit(5))

pprint.pprint(movies_after_2000)
```

```python
high_rated_movies = list(collection.find({"imdb.rating": {"$gt": 8.5}}).limit(5))

pprint.pprint(high_rated_movies)
```

```python
action_adventure_movies = list(collection.find({"genres": {"$all": ["Action",
"Adventure"]}}).limit(5))
pprint.pprint(action_adventure_movies)
```

```python
sorted_comedy_movies = list(collection.find({"genres": "Comedy"}).sort("imdb.rating", -
1).limit(5))
pprint.pprint(sorted_comedy_movies)
```

```python
sorted_drama_movies = list(collection.find({"genres": "Drama"}).sort("year", 1).limit(5))
pprint.pprint(sorted_drama_movies)
```

```python
avg_rating_by_genre = collection.aggregate([
{"$unwind": "$genres"},
{"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
{"$sort": {"avg_rating": -1}},
{"$limit": 5}
])
pprint.pprint(list(avg_rating_by_genre))
```

```python
top_directors = collection.aggregate([
{"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
{"$sort": {"avg_rating": -1}},
{"$limit": 5}
])
pprint.pprint(list(top_directors))
```

```python
movies_per_year = collection.aggregate([
{"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
{"$sort": {"_id": 1}}
])
pprint.pprint(list(movies_per_year))
```

```python
collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})
```

```python
collection.update_many({"genres": "Horror", "imdb.rating": {"$exists": False}}, {"$set":
{"imdb.rating": 6.0}})
```

```python
collection.delete_many({"year": {"$lt": 1950}})
```

```python
print("Existing indexes:")
indexes = collection.list_indexes()
for index in indexes:
    print(f"Index Name: {index['name']}, Key: {index['key']}")
```

```python
collection.create_index([("title", "text")])
love_movies = collection.find({"$text": {"$search": "love"}})
collection.create_index([("title", "text"), ("plot", "text")])
war_movies = collection.find({"$text": {"$search": "war"}}).sort("imdb.rating", -
1).limit(5)
```

```python
action_high_rated_movies = collection.find({"genres": "Action", "imdb.rating": {"$gt":
8}}).sort("year", -1)
pprint.pprint(list(action_high_rated_movies))
```

```python
nolan_movies = collection.find({"directors": "Christopher Nolan", "imdb.rating": {"$gt":
8}}).sort("imdb.rating", -1).limit(3)
pprint.pprint(list(nolan_movies))
```
